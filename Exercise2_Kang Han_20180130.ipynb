# EX-2: Confusion Matrix

```python
from sklearn.datasets import load_digits
```

```python
import pandas as pd
```

```python
digits = load_digits()
```

```python
y = digits.target == 9 #숫자 9를 positive class로 설정
```

```python
pd.DataFrame(digits.target)
```

```python
from sklearn.model_selection import train_test_split
```

```python
pd.DataFrame(digits.data)
```

```python
X_train, X_test, y_train, y_test = train_test_split(
    digits.data, y, random_state=0)
```

```python
pd.DataFrame(X_train)
```

```python
pd.DataFrame(X_test)
```

```python
pd.DataFrame(y_train)
```

```python
pd.DataFrame(y_test)
```

```python
1347+450
```

### Training Models

```python
from sklearn.dummy import DummyClassifier
```

```python
dummy = DummyClassifier(strategy='most_frequent').fit(X_train, y_train)
```

```python
pred_dummy = dummy.predict(X_test)
```

```python
pred_dummy
```

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
tree = DecisionTreeClassifier(max_depth=2).fit(X_train, y_train)
```

```python
pred_tree = tree.predict(X_test)
```

```python
pred_tree
```

### Accuracy

```python
from sklearn.metrics import accuracy_score
```

```python
print("Dummy model:")
print(accuracy_score(y_test, pred_dummy))
print("Decision tree:")
print(accuracy_score(y_test, pred_tree))
```

### Confusion Matrix

```python
from sklearn.metrics import confusion_matrix
```

```python
print("Dummy model:")
print(confusion_matrix(y_test, pred_dummy))
print("Decision tree:")
print(confusion_matrix(y_test, pred_tree))
```

### Classification Report

```python
from sklearn.metrics import classification_report
```

```python
print("Dummy model:")
print(classification_report(y_test, pred_dummy,
                           target_names=["not 9", "9"]))
print("\nDecision tree:")
print(classification_report(y_test, pred_tree,
                           target_names=["not 9", "9"]))
```
