# EX-3: ROC & AUC

```python
from sklearn.metrics import roc_curve
from sklearn.metrics import auc
import matplotlib.pylab as plt
%matplotlib inline
```

##### Define ROC curve drawing function

```python
def plot_roc_curve(fpr, tpr, model, color=None) :
    model = model + ' (auc = %0.3f)' % auc(fpr, tpr)
    plt.plot(fpr, tpr, label=model, color=color)
    plt.plot([0, 1], [0, 1], color='navy', linestyle='--')
    plt.axis([0,1,0,1])
    plt.xlabel('FPR (1 - specificity)')
    plt.ylabel('TPR (recall)')
    plt.title('ROC curve')
    plt.legend(loc="lower right")
```

<font color = "blue"> The named colors used in matplotlib

<font color = "blue">* 수업 자료에서 색상 이름을 참고하라.

##### Plot multiple ROC curves

```python
fpr_dummy, tpr_dummy, _ = roc_curve(y_test,
                                   dummy.predict_proba(X_test)[:,1])
plot_roc_curve(fpr_dummy, tpr_dummy, 'dummy model', 'hotpink')
fpr_tree, tpr_tree, _ = roc_curve(y_test,
                                 tree.predict_proba(X_test)[:,1])
plot_roc_curve(fpr_tree, tpr_tree, 'decision tree', 'darkgreen')
```

# EX-5: Model Tuning

<font color = "blue">* Best parameters를 찾아 전체 training data에 적용하자.

##### Set the parameters for grid search

```python
#param_grid: dictionary with parameters names as keys and lists of parameters settings to try as values.
param_grid = {'C' : [0.001, 0.01, 0.1, 1, 10, 100],
             'gamma': [0.001, 0.01, 0.1, 1, 10, 100]}
param_grid
```

##### Grid search with cross-validation

```python
from sklearn.model_selection import GridSearchCV
from sklearn.svm import SVC
```

```python
grid_search = GridSearchCV(SVC(), param_grid, cv=5, n_jobs=-1)
```

<font color = "blue">
* 위의 코드 가운데 마지막 부분은 'n_jobs=-1'은 코어가 여러 개인 경우, 병렬로 돌려 속도를 빠르게 할 수 있는 것과 관련된 부분이다.
* 튜닝은 매우 오랜 시간이 걸리는 작업이다.
* SVC를 한 번 돌려봐야 한다. 시간을 측정하며, 1 모델을 돌리는데 시간이 얼마나 걸리는지 확인한 다음에 해 봐야 한다.
* 이 간단한 모델로 하는데도 시간이 걸린다.

```python
#grid search is very time-consuming
grid_search.fit(X_train, y_train)
```

##### Evaluate the model with best parameters

```python
grid_search.score(X_test, y_test)
```

<font color = "blue">
* 위의 결과는 toy example이기 때문에 가능한 결과다.

```python
print("Best parameters: {}".format(grid_search.best_params_))
print("Best CV score: {:.2f}".format(grid_search.best_score_))
```

```python
print("Best estimator:\n{}".format(grid_search.best_estimator_))
```
