##### The effect of preprocessing on supervised learning

```python
from sklearn.datasets import load_breast_cancer
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(
    cancer.data, cancer.target, random_state=0)
```

```python
svm = SVC(C=100)
```

```python
svm.fit(X_train, y_train).score(X_test, y_test)
```

다음은 데이터를 SVM에 맞는 0~1로 바꾸기 위한 code다.

```python
#Preprocessing using 0-1 scaling
from sklearn.preprocessing import MinMaxScaler
```

```python
scaler = MinMaxScaler()
```

```python
X_train_scaled = scaler.fit(X_train).transform(X_train)
```

```python
#Scaling training and test data the same way
X_test_scaled = scaler.fit(X_test).transform(X_test)
```

```python
pd.DataFrame(X_train_scaled)
```

```python
pd.DataFrame(X_test)
```

```python
svm.fit(X_train_scaled, y_train).score(X_test_scaled, y_test)
```

# EX-8: Feature Selection

##### Model based feature selection

아래 예제는 default다. 네가 알아서 판단해 달라는 방식이다.

```python
from sklearn.feature_selection import SelectFromModel
```

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
select = SelectFromModel(RandomForestClassifier(), threshold=None)
```

```python
X_train_fs = select.fit(X_train, y_train).transform(X_train)
```

```python
print("X_train.shape: {}, X_train_fs.shape: {}".format(
    X_train.shape, X_train_fs.shape))
```

'X_train.shape: (426, 30), X_train_fs.shape: (426, 6)'라는 위 코드 실행 결과는, Random Forest가 6개 피처가 의미 있다고 골라 줬다는 것이다.

```python
mask = select.get_support()
```

```python
plt.matshow(mask.reshape(1,-1), cmap="gray_r")
```

<font color = "blue"> In numpy.reshape(), one shape dimension can be -1. In this case, the value is inferred from the length of the array and remaining dimensions.

<font color = "blue">All built-in colormaps can be reversed by appending _r: For instance, gray_r is the reverse of gray.

<font color = "blue">See color map(https://matplotlib.org/2.0.2/api/pyplot_summary.html#matplotlib.pyplot.colormaps).

```python
X_test_fs = select.transform(X_test)
```

```python
svm.fit(X_train_fs, y_train).score(X_test_fs, y_test)
```

* 피처 선택 결과, 성능이 떨어졌다.
* 중요하다고 선택된 것 중, 상위 얼마만큼을 할 것인지, 말 것인지 등을 살펴 봐야 한다.
